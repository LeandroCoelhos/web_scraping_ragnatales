```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
pd.set_option("display.max_rows", 50)
```

```python
response = requests.get('https://g1.globo.com/')

content = response.content

site = BeautifulSoup(content, 'html.parser') #conteúdo html da página
```

```python
print(site.prettify()) #imprime o site no formato html identado)
```

```python
# HTML da notícia
noticia = site.find('div', attrs={'class': 'feed-post-body'}) # pegar a primeira marcação, especifico um segundo nível arbitrário
# Título
titulo = noticia.find('a', attrs={'class': 'feed-post-link'})
print(noticia.prettify())# Ajuda a vizualizar o que esta vindo em cada div
```

```python
print(titulo.prettify())# vou descendo o nível
```

```python
print(titulo.text) #pega a parte do texto que não esta entre os < >
```

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep
```

```python
options = Options()
options.add_argument('window-size=800,800') #define o tamanho da janela
#options.add_argument('--headless')# Não abre o visual do navegador
```

```python
url = 'https://www.airbnb.com.br/'
response = requests.get(url)
site = BeautifulSoup(response.text, 'html.parser')
site.prettify()
```

```python
navegador = webdriver.Chrome(options=options)
navegador.get(url)
sleep(2)

input_place = navegador.find_element(by='tag name', value='input')# procura um elemto na página pela tag
```

```python
input_place.send_keys('São Paulo') #escreve no elemento ou campo achado
input_place.submit()# aperta enter ou enviar
```

```python
sleep(0.5)
button_stay = navegador.find_elemt(by='css selector', value='button > img')# um botão e logo depois dentro dela tem uma imagem
button_stay.click() #clica no botão
```

```python
sleep(0.5)
nextButton = navegador.find_element(by='tag name', value='button')[-1]
nextButton.click()
```

```python
sleep(0.5)
searchButton = navegador.find_element(by='tag name', value='button')[-1]
searchButton.click()
```

```python
sleep(0.5)
page_content = navegador.page_source # tira um print do htlm da pagina nesse momento
```

```python
site = BeautifulSoup(page_content, 'html.parser')
site.prettify() #da uma ajeitada no codigo
```
